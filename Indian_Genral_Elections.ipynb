__State wise performance of BJP and Congress__

```python
e1=pd.DataFrame(df['State'][df['Winning Party']=='Bharatiya Janata Party'].groupby(df['State']).count().sort_values(ascending=False))
plt.bar(e1.index,e1['State'])
plt.xticks(rotation=90)
```

```python
e2=pd.DataFrame(df['State'][df['Winning Party']=='Indian National Congress'].groupby(df['State']).count().sort_values(ascending=False))
e2
plt.bar(e2.index,e2['State'])
plt.xticks(rotation=90)
```
